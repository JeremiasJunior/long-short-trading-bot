```python
import init
from analisys.bkt_analisys import *
from analisys.data_analisys import *
from analisys.graph_analisys import *
from analisys.optmizer_analisys import *
from backend.mt5_tools import mt5_singlehistoricaldata
import seaborn as sns
import plotly.express as px

%matplotlib inline

#pd.set_option("display.max_rows", None,#"display.max_columns", None)

import time
import os

timeframe = 5
```

```python
petr3 = mt5_singlehistoricaldata('PETR3', 16408, '2021.04.04', '2021.11.01')
petr4 = mt5_singlehistoricaldata('PETR4', 16408, '2021.04.04', '2021.11.01')

petr3_close = petr3['close'][::-1]
petr4_close = petr4['close'][::-1]

_data = pd.DataFrame({'petr3':petr3_close, 'petr4':petr4_close})
```

```python
from scipy.stats import pearsonr


corr = pearsonr(petr3_close, petr4_close)

regression = sns.scatterplot(data=_data,  x = 'petr3', y = 'petr4')
regression.set_ylabel('petr4')
regression.set_xlabel('petr3')
regression.set_title('pearson corr = {}'.format(corr[0]))
sns.set_style(style="darkgrid")

```
