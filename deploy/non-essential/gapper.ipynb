```python
#usar na data da ultima segunda-feira

def fii_filter(fii_list_file = 'symbols_fiis.csv', filter = 50000):
    
    hist = mt5_historicaldata_s(60, '2021.11.05', '2021.11.13', csv_list= fii_list_file, reframe=False, print_ticker=False)

    key_raw_list = hist.keys()
    key_new_list = []

    for k in key_raw_list:
        if(np.sum(hist[k]['realvolume'])/5 > filter):
            key_new_list.append(k)

    
    return key_new_list
    
def spread(fii_symbol):
    bid, ask = mt5_currentprice(fii_symbol)
    return round(ask-bid,3)

```

```python
hist = fii_filter()
```

```python
fii_list = {'fii':[], 'spread':[], 'bid':[], 'ask':[]}

for i in hist:
    bid, ask = mt5_currentprice(i)

    fii_list['fii'].append(i)
    fii_list['spread'].append(spread(i))
    fii_list['bid'].append(round(bid, 3))
    fii_list['ask'].append(round(ask, 3))

fii_list = pd.DataFrame(fii_list)

```
